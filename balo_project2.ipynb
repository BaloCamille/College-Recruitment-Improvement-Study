```python
#import pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

#read in the cleaned csv file
data = "C:/Users/baloc/OneDrive/Documents/Python/assignment5/cleaned_data.csv"
df = pd.read_csv(data)
```

Finding 1-- Interest based on student gender

```python
#find all the unique values found in the gender column
gender_values = df['gender'].unique()

#the gender_values array that gets returned by the unique() method includes the following three values found:
    #man, woman, and prefer not to say

#find the count of each gender that responded to the survey
man_counts = df['gender'].value_counts()['Man']
woman_counts = df['gender'].value_counts()['Woman']
notsay_counts = df['gender'].value_counts()['Prefer not to say']

#the returned results are:
    #man_counts = 52
    #woman_counts = 77
    #notsay_counts = 2
```

```python
#find the breakdown of males' and females' interest in taking another IT course
#create boolean masks to use later to locate the rows that match the mask

#this mask will find males that responded as highly interested (4 or 5)
high_male_mask = (df['gender'] == 'Man') & (df['again_interest'].isin([4, 5]))
#this mask will find males who are neutral (3)
neut_male_mask = (df['gender'] == 'Man') & (df['again_interest'].isin([3]))
#this mask will find males who are least interested (1 or 2)
low_male_mask = (df['gender'] == 'Man') & (df['again_interest'].isin([1,2]))

#this mask will find females who are highly interested (4 or 5)
high_fem_mask = (df['gender'] == 'Woman') & (df['again_interest'].isin([4,5]))
#this mask will find females who are neutral (3)
neut_fem_mask = (df['gender'] == 'Woman') & (df['again_interest'].isin([3]))
#this mask will find females who are least interested (1 or 2)
low_fem_mask = (df['gender'] == 'Woman') & (df['again_interest'].isin([1,2]))

#find counts of each of the masks using the loc() and len methods
high_male_count = len(df.loc[high_male_mask]) #returned 9
neut_male_count = len(df.loc[neut_male_mask]) #returned 24
low_male_count = len(df.loc[low_male_mask]) #returned 19
high_fem_count = len(df.loc[high_fem_mask]) #returned 10
neut_fem_count = len(df.loc[neut_fem_mask]) #returned 29
low_fem_count = len(df.loc[low_fem_mask]) #returned 38

#create a variable to hold the x_values (interest levels)
x_values = ('Low', 'Neutral', 'High')

#create a list with the 3 male counts
total_male_counts = (low_male_count, neut_male_count, high_male_count)

#plot the findings
plt.bar(x_values, total_male_counts)

#customize and show the plot
plt.xlabel('Interest Level')
plt.ylabel('Count')
plt.title('Interest Level Counts for Males')
plt.show()


#do the same for females now
#create a list with the 3 male counts
total_fem_counts = (low_fem_count, neut_fem_count, high_fem_count)

#plot the findings
plt.bar(x_values, total_fem_counts)

#customize and show the plot
plt.xlabel('Interest Level')
plt.ylabel('Count')
plt.title('Interest Level Counts for Females')
plt.show()
```
